```python
# Remove the columns with 1 unique value
for col in list(df.nunique()[df.nunique() == 1].index):
    del df[col]
```

```python
# Define categorical if a column has 6 or less unique values
categorical_cols = list(df.nunique()[df.nunique() <= 6].index)
```

```python
df['y'] = (df['Attrition'] == 'Yes').astype(int)
```

# Explore numerical columns

```python
# Remove EmployeeID
del df['EmployeeID']
```

```python
# Define categorical if a column has more than 6 unique values
numerical_cols = list(df.nunique()[df.nunique() > 6].index)
numerical_cols
```

```python
def continuous_variables_analysis(x,y,data):
    # Continuous variable
    print(x)

    # Plot propensity by percentile
    variable_quantiles = pd.qcut(data[x], q=10, precision=0, duplicates='drop')
    data.groupby(variable_quantiles).mean()[y].plot()
    plt.show()

    # Distribution plot by target value
    for value in data[y].unique(): 
        df_filtered = data[data[y] == value]

        sns.kdeplot(df_filtered[x], label=value)
    plt.legend()
    plt.show()


for col in numerical_cols:
    continuous_variables_analysis(col,'y',df)
```

# Fit model

```python
import sklearn
import warnings

from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
# from sklearn.neural_network import MLPRegressor

from sklearn.pipeline import make_pipeline

from sklearn.compose import make_column_transformer
from sklearn.compose import TransformedTargetRegressor

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score, confusion_matrix

from joblib import dump, load


from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline

from sklearn.compose import make_column_transformer
from sklearn.compose import TransformedTargetRegressor

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = df[categorical_cols + numerical_cols]
y = df['y']

x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 42)
```

```python
preprocessor = make_column_transformer(
    (OneHotEncoder(drop='if_binary'), categorical_cols,),
    (StandardScaler(), numerical_cols),
    remainder='drop'
)

poli = PolynomialFeatures(degree=2, include_bias=False)

preprocessor_pipeline = make_pipeline(
    preprocessor,
#     poli
)

x_train_t = preprocessor_pipeline.fit_transform(x_train)
```

```python
lr = LogisticRegression()
dtc = tree.DecisionTreeClassifier(max_depth=3)
rfc = RandomForestClassifier(max_depth=3)
gbc = GradientBoostingClassifier()
# knn = KNeighborsClassifier(n_neighbors=10, weights='uniform')
# svm = SVC(kernel='rbf', C=1, gamma='auto')
# nn = MLPRegressor(hidden_layer_sizes=100, activation='identity', max_iter=500)

models = (
    ('LinearClassifier',lr),
    ('DecisionTree',dtc),
    ('RandomForestClassifier',rfc),
    ('GradientBoostingClassifier',gbc)
    # ('KNeighborsClassifier',knn),
    # ('SVM',svm),
    # ('MLPRegressor',nn),
    )

cv_results = []
for name,model in models:
    model_pipeline = make_pipeline(
        preprocessor_pipeline,
        model
    )
  
    cv_result = cross_validate(model_pipeline, x_train, y_train, cv=5, scoring=('accuracy', 'f1', 'roc_auc'), return_train_score=True)

    cols = ['test_accuracy','test_f1','test_roc_auc','train_accuracy','train_f1','train_roc_auc']
    melted_df = pd.melt(pd.DataFrame(cv_result), id_vars=None, value_vars=cols, var_name='metric', value_name='score')
    melted_df['model'] = name
    melted_df['split'] = melted_df['metric'].str.split('_',1).apply(lambda x: x[0])
    melted_df['metric'] = melted_df['metric'].str.split('_',1).apply(lambda x: x[1])

    cv_results.append(melted_df)

cv_results = pd.concat(cv_results).reset_index(drop=True)
```

```python
for metric in cv_results['metric'].unique():
    cond = cv_results['metric'] == metric

    print(metric)
    plt.figure(figsize=(20,10))
    sns.boxplot(x="model", y="score", hue="split", data=cv_results[cond])
    plt.title = metric
    plt.show()
```

# Logistic Regression

```python
model = LogisticRegression()

model_pipeline = make_pipeline(
        preprocessor,
        model
    )

model_pipeline.fit(x_train, y_train)
y_pred = model_pipeline.predict(x_test)
y_pred_prob = model_pipeline.predict_proba(x_test)
```

```python
# Evaluate the test predictions against the actual target values to measure model performance
accuracy_test = accuracy_score(y_test,y_pred)
f1_test = f1_score(y_test,y_pred)
roc_test = roc_auc_score(y_test,y_pred)
precision_test = precision_score(y_test,y_pred)
recall_test = recall_score(y_test,y_pred)

print('Accuracy: ',accuracy_test)
print('F1: ',f1_test)
print('ROC: ',roc_test)
print('Precision: ',precision_test)
print('Recall: ',recall_test)
```
